```sql
--What Invoice is Paid/Not Paid Within a Certain Period
SELECT 'Active' AS Source, InvoiceNumber,
        InvoiceDate, InvoiceTotal
FROM ActiveInvoices
WHERE InvoiceDate >= '04/10/2016'
UNION
SELECT 'Paid' AS Source, InvoiceNumber,
        InvoiceDate, InvoiceTotal
FROM PaidInvoices
WHERE InvoiceDate >= '04/10/2016'
ORDER BY InvoiceTotal DESC;
GO

-- 1st Two CustNames Unique to Custs Compared to Employees 
SELECT 
TOP 2 CustomerFirst, CustomerLast
    FROM Customers
EXCEPT
SELECT FirstName, LastName
    FROM Employees
ORDER BY CustomerLast;
GO

-- InvestmentTotal in relation to InvestorID Count
SELECT InvestorID, COUNT(*) AS InvestorCount,
    SUM(Investments) AS InvestmentTotal
FROM Investors
GROUP BY CUBE(InvestorID);
```
